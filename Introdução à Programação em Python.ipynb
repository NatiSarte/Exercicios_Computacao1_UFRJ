```python
# Exercício 01 - Faça um programa que leia um inteiro e mostre na tela seu sucessor e predecessor.

numero = int(input('Digite um número:'))

print(f'O predecessor é {numero -1} e o sucessor {numero +1} desse número')
```

```python
# Exercício 02 - Crie um algoritmo que leia um número e mostre o seu dobro, triplo e raiz quadrada.
numero = int(input('Digite um número: '))

print(f'O dobro, triplo e a raiz quadrada desse número são respectivamente {numero * 2}, {numero * 3} e {numero **(1/2)}')
```

```python
# Exercício 03 - Desenvolva um programa que leia as duas notas de um aluno,
# calcule e mostre a sua média.

primeira_nota = float(input('Primeira nota: '))
segunda_nota = float(input('Segunda nota: '))
soma = primeira_nota + segunda_nota

print(f'A média é {soma/2}')
```

```python
# Exercício 04 - Crie um algoritmo que leia o raio interno e o raio externo,
# calcule e mostre a área da coroa circular. (Pi = 3.14)

raio_interno = float(input('Raio interno: '))
raio_externo = float(input('Raio externo: '))
pi = 3.14
area_interna = ((raio_interno**2)*pi)
area_externa = ((raio_externo**2)*pi)


print(f' A coroa circular tem área igual a {area_externa - area_interna}')
```

```python
# Exercício 05 - Escreva um programa que leia um valor em metros
# e o exiba convertido em centímetros e milímetros.

numero = float(input("Digite um número em metros: "))

print(f'O número pode ser escrito {numero *100}cm ou {numero *1000}mm')
```

```python
# Exercício 06 - Crie um algoritmo que leia uma temperatura em graus Celsius (°C),
# converta e mostra a temperatura em Fahrenheit (°F).

temperatura = float(input('Qual a temperatura? '))

print(f'A temperatura em Fahrenheit é {(temperatura * 1.8) + 32}ºF')
```

```python
# Exercício 07 - Desenvolva um código que dado o valor da conta de um restaurante,
# calcule a gorjeta do garçom, considerando que a gorjeta deve ser 15% do valor da conta.

conta = float(input('Valor da conta: R$'))

print(f'A corjeta do garçom é R${conta * 0.15}')
```

```python
# Exercício 08 - Crie um algoritmo que leia o valor da conta e os dias de atraso,
# calcule e mostre o valor total a pagar. Considere a multa de 0.5% ao dia.

conta = float(input('Valor da conta: R$'))
dias_atraso = int(input('Dias em atraso: '))

print(f'O valor da multa é de: R${conta*0.05*dias_atraso}')
```

```python
# Exercício 09 - Escreva um programa que pergunte a quantidade de Km percorridos por um carro alugado
# e a quantidade de dias pelos quais ele foi alugado. Calcule o preço a pagar,
# sabendo que o carro custa R$60 por dia e R$0,15 por Km rodado.

kilometro = float(input('Kilometros percorridos: '))
dias_usado = int(input('Dias alugado: '))

print(f'Deve pagar o total de R${(60*dias_usado) + (0.15*kilometro)}')
```

```python
# Exercício 10 - Faça um programa que leia o comprimento do cateto oposto
# e do cateto adjacente de um triângulo retângulo. Calcule e mostre o comprimento da hipotenusa.

cateto_oposto = float(input('Valor do cateto oposto: '))
cateto_adjacente = float(input('Valor do cateto adjacente: '))
aux = (cateto_oposto**2) + (cateto_adjacente**2)
print(f'A hipotenusa vale: {aux**(1/2)}')
```
